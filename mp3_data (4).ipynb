```python
with open('data_mp3.json', 'r', encoding='utf-8') as f:
    json_data = json.load(f)

sorular_list = []
for soru in json_data['sorular']:
    for secenek in soru['secenekler']:
        sorular_list.append({
            'soru_id': soru['id'],
            'soru_metni': soru['soru'],
            'cevap': secenek['cevap'],
            'oy': secenek['oy'],
            'yuzde': secenek['yuzde']
        })

data = pd.DataFrame(sorular_list)

print("TOPLAM KAYIT:", len(data))
print("SORU SAYISI:", data['soru_id'].nunique())

for soru_id in sorted(data['soru_id'].unique()):
    soru_data = data[data['soru_id'] == soru_id]
    print(f"\n--- SORU {soru_id} ---")
    print(soru_data[['cevap', 'oy', 'yuzde']].to_string(index=False))
```
